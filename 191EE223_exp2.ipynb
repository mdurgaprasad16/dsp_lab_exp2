```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

from math import pi
from scipy import signal as sp
import IPython
from scipy.io.wavfile import write
from scipy.io import wavfile
import wave
```

```python
#regestration number : 191EE223
num=223
a=1+num%4
pi=np.pi
```

```python
t=np.linspace(-1/(2*a),1/(2*a),1000)
```

```python
#PROBLEM 1.1

cosine_1=a*np.cos(2*pi*t*5*a)
cosine_2=(a/2)*np.cos(2*pi*t*6*a)
cosine_3=(a/4)*np.cos(2*pi*t*10*a)
```

```python
plt.plot(t,cosine_1, label='cosine_1' )
plt.plot(t,cosine_2,label='cosine_2')
plt.plot(t,cosine_3,label='cosine_3')
plt.xlabel('<-----Time----->')
plt.ylabel('( Cosine functions )')
plt.title('Problem 1.1')
plt.legend()
plt.savefig('Problem 1.1.png')
plt.show()
```

```python
#PROBLEM 1.2

sum_1=cosine_1+cosine_2+cosine_3
plt.plot(t,sum_1)
plt.xlabel('Time---->')
plt.ylabel('( Summation of 3 cosines )')
plt.title('Problem 1.2')
plt.savefig('Problem 1.2.png')
plt.show()
```

```python
#PROBLEM 1.3a

fig,ax=plt.subplots(3,1,figsize=(10,10))

Fs_a=14*a
signal_a=sp.resample(sum_1,Fs_a)
plt.subplot(3,1,1)
plt.xlabel('Samples---->')
plt.title(' 1.3 a')
plt.ylabel('( At Fs=14*alpha )')
plt.stem(signal_a)

#PROBLEM 1.3b
Fs_b=20*a
signal_b=sp.resample(sum_1,Fs_b)
plt.subplot(3,1,2)
plt.xlabel('Samples---->')
plt.title(' 1.3 b')
plt.ylabel('( At Fs=20*alpha )')
plt.stem(signal_b)

#PROBLEM 1.3c
Fs_c=9*a
signal_c=sp.resample(sum_1,Fs_c)
plt.subplot(3,1,3)
plt.xlabel('Samples----->')
plt.title(' 1.3 c')
plt.ylabel('( At Fs=9*alpha )')
plt.stem(signal_c)

plt.subplots_adjust(left=0.1,bottom=0.1,right=0.9,top=0.9,wspace=1,hspace=0.5)
plt.savefig('Problem 1.3.png')
plt.show()
```

```python
#PROBLEM 1.4a

fig,ax=plt.subplots(3,1,figsize=(10,10))

Fs_a=14*a
signal_a=sp.resample(sum_1,Fs_a)
plt.subplot(3,1,1)
plt.xlabel('Samples----->')
plt.title(' 1.4 a')
plt.ylabel('( At Fs=14*alpha )')
plt.plot(signal_a)

#PROBLEM 1.4b
Fs_b=20*a
signal_b=sp.resample(sum_1,Fs_b)
plt.subplot(3,1,2)
plt.xlabel('Samples----->')
plt.title(' 1.4 b')
plt.ylabel('( At Fs=20*alpha }')
plt.plot(signal_b)

#PROBLEM 1.4c
Fs_c=9*a
signal_c=sp.resample(sum_1,Fs_c)
plt.subplot(3,1,3)
plt.xlabel('Samples----->')
plt.title(' 1.4 c')
plt.ylabel('( At Fs=9*alpha )')
plt.plot(signal_c)

plt.subplots_adjust(left=0.1,bottom=0.1,right=0.9,top=0.9,wspace=1,hspace=0.5)
plt.savefig('Problem 1.4.png')
plt.show()
```

```python
fig,ax=plt.subplots(3,1,figsize=(10,10))
plt.subplot(3,1,1)
plt.magnitude_spectrum(signal_a)
plt.title(' 1.5 a')
plt.ylabel('( At Fs=14*alpha )')


plt.subplot(312)
plt.magnitude_spectrum(signal_b)
plt.title(' 1.5 b')
plt.ylabel('( At Fs=20*alpha )')


plt.subplot(313)
plt.magnitude_spectrum(signal_c)
plt.title(' 1.5 c')
plt.ylabel('( At Fs=9*alpha )')


plt.subplots_adjust(left=0.1,bottom=0.1,right=0.9,top=0.9,wspace=1,hspace=0.5)
plt.savefig('Problem 1.5.png')
plt.show()
```

```python
sampling_rate=500
```

```python
t2=np.linspace(0,1,sampling_rate)
Do=np.sin(2*np.pi*180*t2)
Re=np.sin(2*np.pi*200*t2)
Mi=np.sin(2*np.pi*220*t2)
Fa=np.sin(2*np.pi*250*t2)
So=np.sin(2*np.pi*260*t2)
La=np.sin(2*np.pi*300*t2)
Ti=np.sin(2*np.pi*300*t2)
Doo=np.sin(2*np.pi*320*t2)
```

```python
DoRe=np.append(Do,Re,axis=0)
DoReMi=np.append(DoRe,Mi,axis=0)
DoReMiFa=np.append(DoReMi,Fa,axis=0)
DoReMiFaSo=np.append(DoReMiFa,So,axis=0)
DoReMiFaSoLa=np.append(DoReMiFaSo,La,axis=0)
DoReMiFaSoLaTi=np.append(DoReMiFaSoLa,Ti,axis=0)
DoReMiFaSoLaTiDoo=np.append(DoReMiFaSoLaTi,Doo,axis=0)
```

```python
write("sampled_500.wav", sampling_rate, DoReMiFaSoLaTiDoo.astype(np.float32))
IPython.display.Audio(DoReMiFaSoLaTiDoo,rate=sampling_rate)
```

```python
#Upsampling
fname = 'Track004.wav'
rate, data = wavfile.read('Track004.wav')

# resample using FFT and convert back to int16
resampling_factor = 2
samples = len(data) * resampling_factor
rate_new = rate * resampling_factor
resampled_data = sp.resample(data, samples).astype(np.int16)

# append data to ndarray - add the last sample
n = len(resampled_data)
value = resampled_data[-1]
resampled_data_1 = np.append(resampled_data, value)
n = len(resampled_data_1)

# write wav file
fname_new = fname + '.new'
write("upsampled2.wav", rate_new, resampled_data_1)
```
